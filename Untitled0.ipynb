```python
!pip install opencv-python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
image = cv2.imread(list(uploaded.keys())[0])
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
sampled_image = cv2.resize(image, (image.shape[1]//4, image.shape[0]//4))

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(sampled_image)
plt.title("Sampled Image")
plt.axis("off")

plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

def quantize_image(img, levels):
    step = 256 // levels
    quantized = (img // step) * step
    return quantized

quantized_16 = quantize_image(gray, 16)
quantized_4 = quantize_image(gray, 4)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(quantized_16, cmap='gray')
plt.title("16 Levels")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(quantized_4, cmap='gray')
plt.title("4 Levels")
plt.axis("off")

plt.show()
```
